```python
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

%matplotlib inline

def create_plot(file):
    data = open(file).read().strip().split("\n")

    epochs = []
    results = []

    for row in data:
        epoch, result = row.split(',')
        epochs.append(int(epoch))
        results.append(float(result))

    for result in results:
        result = round(result, 3)

    plt.plot(epochs, results, 'ko', markersize=2)
    return plt.show()

print(1)
create_plot("data/e5000l09m00") # learning .9 momentum .0
print(2)
create_plot("data/e5000l06m00") # learning .6 momentum .0
print(3)
create_plot("data/e5000l02m00") # learning .2 momentum .0
print(4)
create_plot("data/e5000l09m06") # learning .9 momentum .6
print(5)
create_plot("data/e5000l02m09") # learning .2 momentum .9
print(6)
create_plot("data/e5000l09m00b") # learning .9 momentum .0 + bias
print(7)
create_plot("data/e5000l06m00b") # learning .6 momentum .0 + bias
print(8)
create_plot("data/e5000l02m00b") # learning .2 momentum .0 + bias
print(9)
create_plot("data/e5000l09m06b") # learning .9 momentum .6 + bias
print(10)
create_plot("data/e5000l02m09b") # learning .2 momentum .9 + bias
print(11)
create_plot("data/e100000l02m09") # learning .2 momentum .9 + bias
```
