```python
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')

%matplotlib inline

def create_plot(file):
    data = open(file).read().strip().split("\n")

    epochs = []
    results = []

    for row in data[1:-1]:
        epoch, result = row.split(',')
        epochs.append(int(epoch))
        results.append(float(result))

    for result in results:
        result = round(result, 3)

    plt.plot(epochs, results, 'ko', markersize=2)
    return plt.show()


create_plot("data/iris433_120000")
create_plot("data/iris443_120000")
create_plot("data/iris4343_40000")
create_plot("data/iris473_100000")
create_plot("data/iris4433_40000")
```
